```python
import torch
import whisper
import time
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from jiwer import wer, cer
from datasets import load_dataset, Audio
from transformers import pipeline
from tqdm import tqdm
import torchaudio
import numpy as np
from nemo.collections.asr.models import ASRModel

# ✅ Authenticate with Hugging Face
hf_token = "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"

# ✅ Load Ionio-ai/speech_to_text_benchmark dataset
print("🔽 Loading Ionio-ai/speech_to_text_benchmark dataset...")
dataset = load_dataset(
    "Ionio-ai/speech_to_text_benchmark",
    split="train",
    token=hf_token
).cast_column("audio", Audio(sampling_rate=16000))

# ✅ Function to add noise for robustness testing
def add_noise(audio_array, noise_level=0.01):
    noise = np.random.normal(0, noise_level, len(audio_array))
    return audio_array + noise

# ✅ Load ASR Models
print("⏳ Loading ASR models...")
models = {
    "Whisper Base": whisper.load_model("base"),
    "Whisper Medium": whisper.load_model("medium"),
    "Whisper Large": whisper.load_model("large"),
    "Wav2Vec2": pipeline(
        "automatic-speech-recognition",
        model="facebook/wav2vec2-large-960h",
        device=0 if torch.cuda.is_available() else -1
    ),
    "HuBERT": pipeline(
        "automatic-speech-recognition",
        model="facebook/hubert-large-ls960-ft",
        device=0 if torch.cuda.is_available() else -1
    ),
    "NeMo Conformer": ASRModel.from_pretrained("stt_en_conformer_ctc_large")
}

# ✅ Evaluate
print("\n🚀 Starting evaluation...\n")
results = []

for model_name, model in models.items():
    print(f"🔍 Evaluating: {model_name}")
    for sample in tqdm(dataset):
        reference = sample["text"]
        audio_array = sample["audio"]["array"]
        sampling_rate = sample["audio"]["sampling_rate"]
        audio_duration = len(audio_array) / sampling_rate

        # Clean audio evaluation
        start = time.time()
        try:
            if "Whisper" in model_name:
                pred = model.transcribe(audio_array.astype("float32"))["text"]
            elif "NeMo" in model_name:
                audio_tensor = torch.tensor(audio_array).float().unsqueeze(0)
                pred = model.transcribe([audio_tensor])[0]
            else:
                pred = model({"array": audio_array, "sampling_rate": sampling_rate})["text"]
        except Exception as e:
            pred = f"[ERROR: {e}]"
        end = time.time()

        latency = round(end - start, 2)
        rtf = round(latency / audio_duration, 3) if audio_duration > 0 else None
        wer_score = round(wer(reference.lower(), pred.lower()), 3) if not pred.startswith("[ERROR") else None
        cer_score = round(cer(reference.lower(), pred.lower()), 3) if not pred.startswith("[ERROR") else None

        results.append({
            "Model": model_name,
            "Reference": reference,
            "Prediction": pred,
            "WER": wer_score,
            "CER": cer_score,
            "Latency (s)": latency,
            "RTF": rtf,
            "Condition": "Clean"
        })

        # Noisy audio evaluation (10 dB SNR equivalent)
        noisy_audio = add_noise(audio_array, noise_level=0.01)
        start = time.time()
        try:
            if "Whisper" in model_name:
                pred_noisy = model.transcribe(noisy_audio.astype("float32"))["text"]
            elif "NeMo" in model_name:
                audio_tensor = torch.tensor(noisy_audio).float().unsqueeze(0)
                pred_noisy = model.transcribe([audio_tensor])[0]
            else:
                pred_noisy = model({"array": noisy_audio, "sampling_rate": sampling_rate})["text"]
        except Exception as e:
            pred_noisy = f"[ERROR: {e}]"
        end = time.time()

        latency_noisy = round(end - start, 2)
        rtf_noisy = round(latency_noisy / audio_duration, 3) if audio_duration > 0 else None
        wer_noisy = round(wer(reference.lower(), pred_noisy.lower()), 3) if not pred_noisy.startswith("[ERROR") else None
        cer_noisy = round(cer(reference.lower(), pred_noisy.lower()), 3) if not pred_noisy.startswith("[ERROR") else None

        results.append({
            "Model": model_name,
            "Reference": reference,
            "Prediction": pred_noisy,
            "WER": wer_noisy,
            "CER": cer_noisy,
            "Latency (s)": latency_noisy,
            "RTF": rtf_noisy,
            "Condition": "Noisy"
        })

# ✅ Summary
df = pd.DataFrame(results)
summary = df.groupby(["Model", "Condition"])[["WER", "CER", "Latency (s)", "RTF"]].mean().reset_index()

print("\n📊 Average Metrics per Model and Condition:")
print(summary)

print("\n📋 Sample Transcriptions:")
print(df.head(10))

# ✅ Visualize Results
plt.figure(figsize=(12, 6))
sns.barplot(data=summary, x="Model", y="WER", hue="Condition")
plt.title("WER Comparison Across Models (Clean vs. Noisy)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(data=summary, x="Model", y="Latency (s)", hue="Condition")
plt.title("Latency Comparison Across Models (Clean vs. Noisy)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(data=summary, x="Model", y="RTF", hue="Condition")
plt.title("Real-Time Factor Comparison Across Models (Clean vs. Noisy)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
